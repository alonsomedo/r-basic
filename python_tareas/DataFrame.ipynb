# Manipulacin de data frames

```python
df = pd.DataFrame(
    {
        'key':['A', 'B', 'C','A', 'B', 'C'],
        'values1':range(6),
        'values2':np.random.randint(0,100,6)
    }
)
```

```python
df
```

```python
df.groupby('key').aggregate([np.mean, np.median, min , max])
```

```python
df.groupby('key').aggregate({
    'values1':'min',
    'values2': [max,np.mean]
})
```

# Filter


```python
df.groupby('key').std()
```

```python
def filter_func(x):
    return x['values2'].std() > 5
```

```python
display('df',"df.groupby('key').std()","df.groupby('key').filter(filter_func)")
```

```python
df.groupby('key').filter(filter_func) #Devuelve las filas que cumplan la condicin
```

```python
df[df.values2>80]
```

```python
df['values2']>80
```

# Transform

```python
df.groupby('key').transform(lambda x: x- x.mean())
```

## Apply

```python
def norm_by_col2(x):
    x['values1']/= x['values2'].sum()
    return x
```

```python
df.groupby('key').apply(norm_by_col2)
```

## Graficos desde data frames y exportacion

```python
years = [year for year in range(1990,2020)] #esta formado por cada uno de los entre 1990 y 2020
deads = [(y + np.random.uniform(0,100)-1850) for y in years]
```

```python
years[0:10]
```

```python
deads[0:5]
```

```python
df = pd.DataFrame({
    'year':years,
    'deads':deads
})
```

```python
df.head()
```

```python
df.plot(x="year",y="deads",figsize = (10,8), kind="scatter")
```

```python
my_plot = df.plot(x="year",y="deads",figsize = (10,8))
```

```python
my_fig = my_plot.get_figure()
```

```python
my_fig.savefig("muertes.png")
```
